```python
import os
import torch
from torch import nn
from torchvision.datasets import MNIST
from torch.utils.data import DataLoader
from torchvision import transforms
```

```python
class MLP(nn.Module):
    '''
    Multilayer Perceptron.
    '''
    def __init__(self):
        super().__init__()
        self.layers = nn.Sequential(
          nn.Flatten(),
          nn.Linear(28 * 28 * 1, 256),
          nn.ReLU(),
          nn.Linear(256, 128),
          nn.ReLU(),
          nn.Linear(128, 64),
          nn.ReLU(),
          nn.Linear(64, 10)
        )


    def forward(self, x):
        '''Forward pass'''
        return self.layers(x)

    def compute_l1_loss(self, w):
        return torch.abs(w).sum()

    def compute_l2_loss(self, w):
        return torch.square(w).sum()
```

```python
if __name__ == '__main__':

    # Check if GPU is available
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    print(f"Training on {device}")

    # Set fixed random number seed
    torch.manual_seed(42)
    
    epochs = 50

    # Prepare MNIST dataset
    train_dataset = MNIST(os.getcwd(), train=True, download=True, transform=transforms.ToTensor())
    test_dataset = MNIST(os.getcwd(), train=False, download=True, transform=transforms.ToTensor())
    
    trainloader = torch.utils.data.DataLoader(train_dataset, batch_size=256, shuffle=True, num_workers=1)
    testloader = torch.utils.data.DataLoader(test_dataset, batch_size=256, shuffle=False, num_workers=1)

    # Initialize the MLP
    mlp = MLP().to(device)
    
    linear_layers = []
    accuracies= []

    for layer in mlp.layers:
        if(isinstance(layer, nn.Linear)):
            linear_layers.append(layer)

    # Define the loss function and optimizer
    loss_function = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(mlp.parameters(), lr=1e-3)

    # Run the training loop
    for epoch in range(epochs): # 5 epochs at maximum

        # Print epoch
        print(f'Starting epoch {epoch+1}')

        # Iterate over the DataLoader for training data
        for i, data in enumerate(trainloader, 0):

            # Get inputs
            inputs, targets = data
            inputs, targets = inputs.to(device), targets.to(device)

            # Zero the gradients
            optimizer.zero_grad()

            # Perform forward pass
            outputs = mlp(inputs)

            # Compute loss
            loss = loss_function(outputs, targets)

            # Specify L1 and L2 weights
            l1_weight = 0#0.0001
            l2_weight = 0#0.0001

            # Compute L1 and L2 loss component
            parameters = []
            for parameter in mlp.parameters():
                parameters.append(parameter.view(-1))
            l1 = l1_weight * mlp.compute_l1_loss(torch.cat(parameters))
            l2 = l2_weight * mlp.compute_l2_loss(torch.cat(parameters))

            # Add L1 and L2 loss components
            loss += l1
            loss += l2

            # Perform backward pass
            loss.backward()

            # Perform optimization
            optimizer.step()

            # Print statistics
            minibatch_loss = loss.item()
            if i % 5000 == 4999:
                print('Loss after mini-batch %5d: %.5f (of which %.5f L1 loss; %0.5f L2 loss)' %
                    (i + 1, minibatch_loss, l1, l2))
                current_loss = 0.0
                
        for count, linear in enumerate(linear_layers):
            if count==0 or count == len(linear_layers)-1:
                continue
            inc = linear_layers[count-1].weight
            out = mlp.layers[2](linear.weight)
            shape = linear_layers[count-1].weight.shape[0]

            for i in range(shape):
                incoming = torch.linalg.norm(linear_layers[count-1].weight[i], dim=None, ord=2)
                outgoing = torch.linalg.norm(mlp.layers[2](linear.weight)[:,i], dim=None, ord=2)
                optimal_l = torch.sqrt(outgoing/incoming)
                linear_layers[count-1].weight[i].data *= optimal_l
                mlp.layers[2](linear.weight)[:,i].data /= optimal_l

        # Testing the model
        correct = 0
        total = 0
        with torch.no_grad():
            for data in testloader:
                images, labels = data
                images, labels = images.to(device), labels.to(device)
                outputs = mlp(images)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        print('Accuracy of the network on the 10000 test images: {:.2f} %'.format(100 * correct / total))
        accuracies.append(100 * correct / total)
```
