```python
import os
import torch
from torch import nn
from torchvision.datasets import MNIST
from torch.utils.data import DataLoader
from torchvision import transforms

from models import *
```

```python
class CustomLinear(nn.Module):
    def __init__(self, input_features, output_features, bias=True):
        super(CustomLinear, self).__init__()
        # Use the nn.Linear layer internally
        self.linear = nn.Linear(input_features, output_features, bias)
        
    @property
    def weight(self):
        return self.linear.weight

    def forward(self, input):
        # Forward pass through the internal nn.Linear layer
        return self.linear(input)
    
    def neural_balance(self, previous_layer, return_norms = False, order = 2, random_perc = 0.1):
        shape = previous_layer.weight.shape[0]
        norm = []
        
#         incoming = torch.linalg.norm(previous_layer.weight, dim=1, ord=2)
#         outgoing = torch.linalg.norm(self.linear.weight, dim=0, ord=2)
#         optimal_l = torch.sqrt(outgoing/incoming)
#         print(optimal_l)
#         previous_layer.weight.data *= optimal_l.unsqueeze(1)
#         self.linear.weight.data /= optimal_l
#         if return_norms:
#             norm.append(torch.linalg.norm(incoming/outgoing, dim = 0, ord=2))
#         if return_norms: return torch.mean(torch.stack(norm))
#         else: return torch.tensor([])
        for i in range(shape):
            incoming = torch.linalg.norm(previous_layer.weight[i], dim=None, ord=order)
            outgoing = torch.linalg.norm(self.linear.weight[:,i], dim=None, ord=order)
            optimal_l = torch.sqrt(outgoing/incoming)
#             print(optimal_l)
            previous_layer.weight[i].data *= optimal_l
            self.linear.weight[:,i].data /= optimal_l
            if return_norms:
                norm.append(torch.linalg.norm(incoming/outgoing, dim = 0, ord=order))
        if return_norms: return torch.mean(torch.stack(norm))
        else: return torch.tensor([])
```

```python
class MLP(nn.Module):
    '''
    Multilayer Perceptron.
    '''
    def __init__(self):
        super().__init__()
        self.layers = nn.Sequential(
          nn.Flatten(),
          CustomLinear(28 * 28 * 1, 256),
          nn.ReLU(),
#           nn.Linear(256, 256),
#           nn.ReLU(),
          CustomLinear(256, 128),
          nn.ReLU(),
          CustomLinear(128, 64),
          nn.ReLU(),
#           nn.Linear(64, 64),
#           nn.ReLU(),
          CustomLinear(64, 10)
        )


    def forward(self, x):
        '''Forward pass'''
        return self.layers(x)

    def compute_l1_loss(self, w):
        return torch.abs(w).sum()

    def compute_l2_loss(self, w):
        return torch.square(w).sum()
```

## training with neural balance

```python
if __name__ == '__main__':

    # Check if GPU is available
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    print(f"Training on {device}")

    # Set fixed random number seed
    torch.manual_seed(42)
    
    epochs = 20
    order = 1

    # Prepare MNIST dataset
    train_dataset = MNIST(os.getcwd(), train=True, download=True, transform=transforms.ToTensor())
    test_dataset = MNIST(os.getcwd(), train=False, download=True, transform=transforms.ToTensor())
    
    trainloader = torch.utils.data.DataLoader(train_dataset, batch_size=256, shuffle=True, num_workers=1)
    testloader = torch.utils.data.DataLoader(test_dataset, batch_size=256, shuffle=False, num_workers=1)

    # Initialize the MLP
    mlp = MLP().to(device)
    
    linear_layers = []
    accuracies= []
    norms = []

    for layer in mlp.layers:
        if(isinstance(layer, CustomLinear)):
            linear_layers.append(layer)

    # Define the loss function and optimizer
    loss_function = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(mlp.parameters(), lr=1e-3)

    # Run the training loop
    for epoch in range(epochs): # 5 epochs at maximum

        # Print epoch
        print(f'Starting epoch {epoch+1}')

        # Iterate over the DataLoader for training data
        for i, data in enumerate(trainloader, 0):

            # Get inputs
            inputs, targets = data
            inputs, targets = inputs.to(device), targets.to(device)

            # Zero the gradients
            optimizer.zero_grad()

            # Perform forward pass
            outputs = mlp(inputs)

            # Compute loss
            loss = loss_function(outputs, targets)

            # Specify L1 and L2 weights
            l1_weight = 0#0.0001
            l2_weight = 0#0.0001

            # Compute L1 and L2 loss component
            parameters = []
            for parameter in mlp.parameters():
                parameters.append(parameter.view(-1))
            l1 = l1_weight * mlp.compute_l1_loss(torch.cat(parameters))
            l2 = l2_weight * mlp.compute_l2_loss(torch.cat(parameters))

            # Add L1 and L2 loss components
            loss += l1
            loss += l2

            # Perform backward pass
            loss.backward()

            # Perform optimization
            optimizer.step()

            # Print statistics
            minibatch_loss = loss.item()
            if i % 5000 == 4999:
                print('Loss after mini-batch %5d: %.5f (of which %.5f L1 loss; %0.5f L2 loss)' %
                    (i + 1, minibatch_loss, l1, l2))
                current_loss = 0.0
        if epoch%1==0:
            for count, linear in enumerate(linear_layers):
                if count==0 or count == len(linear_layers)-1:
                    continue

                norm = linear.neural_balance(linear_layers[count-1], order = order, return_norms=True)
                norms.append(norm)
            
        # Testing the model
        correct = 0
        total = 0
        with torch.no_grad():
            for data in testloader:
                images, labels = data
                images, labels = images.to(device), labels.to(device)
                outputs = mlp(images)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        print('Accuracy of the network on the 10000 test images: {:.2f} %'.format(100 * correct / total))
        accuracies.append(100 * correct / total)
```
