```python
import pickle
import matplotlib.pyplot as plt

with open('/Users/iandomingo/Desktop/code/Neural-Balance/exps/cleanAcc.pkl', 'rb') as f:
    cleanAcc = pickle.load(f)

with open('/Users/iandomingo/Desktop/code/Neural-Balance/exps/nbOnlineEveryEpochAcc.pkl', 'rb') as f:
    nbOnlineEveryEpochAcc = pickle.load(f)

epochs_plot = list(range(100))

plt.figure(figsize=(16, 10))
plt.plot(epochs_plot, cleanAcc, linestyle='-', color='b', label='CleanAcc')
plt.plot(epochs_plot, nbOnlineEveryEpochAcc, linestyle='-', color='r', label='online neural balancing every epoch')

plt.xlabel('Epochs', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.title('Model Accuracy Over Epochs', fontsize=14)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.grid(True)

plt.legend(fontsize=14)

plt.show()
```
